```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import Ridge, Lasso
```

```python
# Import dataset

# Step 1: Load the dataset
# Using the California Housing dataset for regression
housing = datasets.fetch_california_housing()
X = housing.data     # Features
y = housing.target   # Target variable
df = pd.DataFrame(X, columns=housing.feature_names)
df['Median House value'] = y
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.info()
```

```python
df.describe()
```

```python
X_train, X_test, y_train,y_test =  train_test_split(X, y, test_size = 0.25)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# y_train = y_train.reshape(-1,1)
# print(y_train.shape)
```

```python
# y_test = y_train.reshape(-1,1)
# print(y_test.shape)
```

```python
ridge_model = Ridge(alpha = 1.0)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
y_pred_ridge = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, mean_squared_error,r2_score

mse = mean_squared_error(y_test, y_pred_ridge)
mae = mean_absolute_error(y_test, y_pred_ridge)
rmse = root_mean_squared_error(y_test, y_pred_ridge)
r2s = r2_score(y_test, y_pred_ridge)
print(mse, mae, rmse, r2s)
```

```python
print(y_test.shape, y_pred_ridge.shape)
```

```python
lasso_model = Lasso(alpha = 1)
```

```python
lasso_model.fit(X_train, y_train)
```

```python
y_pred_lasso = lasso_model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, mean_squared_error,r2_score

mse = mean_squared_error(y_test, y_pred_lasso)
mae = mean_absolute_error(y_test, y_pred_lasso)
rmse = root_mean_squared_error(y_test, y_pred_lasso)
r2s = r2_score(y_test, y_pred_lasso)
print(mse, mae, rmse, r2s)
```

```python
pd.DataFrame(np.c_[y_test, y_pred_ridge, y_pred_lasso], columns = ['Target Actual', 'Target Predicted(Ridge)', 'Target Predicted(Lasso)'])
```
